# Customer Dempgraphics Analysis

```python
#What is the distribution of customer ages,income, marital situation, education, and other demographic factors?
import matplotlib.pyplot as plt
import seaborn as sns


# lets analyse de age distribution 

print(data['Age'].describe())

# lets visualize the Age distribution in our dataset 

plt.figure(figsize=(7,4))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

The age distribution in our dataset provides insights into the demographic structure of our clients. it reveals that the age range spans from 40 to 50 years is the dominant with a mean of 51 years old 
Our histogram shows that the data is right-skewed, meaning the majority of individuals aged more than 40 years old 

```python
#lets do the same thing with income and education 
print(data['Income'].describe())

# lets plot the income distribution 

plt.figure(figsize=(8,5))
sns.histplot(data['Income'], bins=30, kde=True)
plt.title("Income distribution")
plt.xlabel("Income")
plt.ylabel("frequency")
plt.show()
```

The income distribution in our dataset insights the financial situation of our clients. The mean is 51622 dollars and the median is 51287 dollars which indicates that the distribution is symmetric and normal 
we can say that our dataset is homogenous regarding the financial situation of our clients 

```python
# the education level and marital situation and also the home composition 

# Create two sublots to plot the marital situation and the education level 
fig,(ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18,6))

# the education level distribution 
education_columns = ['education_2n Cycle','education_Basic', 'education_Graduation', 'education_Master', 'education_PhD']
education_counts = data[education_columns].sum()
    
    # plot the distribution 
sns.barplot(x= education_counts.index, y=education_counts.values, ax=ax1, palette='deep',hue = education_counts.values )
ax1.set_title("Education level distribution")
ax1.set_ylabel("Count")
ax1.set_xlabel("Education Level")
ax1.tick_params(rotation= 45)


# the marital situation distribution 
marital_columns = ["marital_Divorced",  "marital_Married", "marital_Single", "marital_Together", "marital_Widow"]
marital_counts = data[marital_columns].sum()

    #plot the marital distribution 
sns.barplot(x=marital_counts.index, y=marital_counts.values, ax=ax2, palette='rocket',hue = marital_counts.values)
ax2.set_title("Marital Situation Distribution")
ax2.set_ylabel("Count")
ax2.set_xlabel("Marital Situation")
ax2.tick_params(rotation = 45)

# the Home composition situation distribution
home_situation = ['Kidhome', 'Teenhome']
home_counts = data[home_situation].sum()

    # plot the home situation 
sns.barplot(x= home_counts.index, y=home_counts.values, ax=ax3, palette='deep', hue= home_counts.values)
ax3.set_title("Home Composition Distribution")
ax3.set_ylabel("Count")
ax3.set_xlabel("Home composition")
ax3.tick_params(rotation = 45)


#show the subplots 
plt.show()



```

The Education level and marital situation distribution in our dataset gives us more importnt insights about the demographic structure of our clients. This analysis reveals that the big majority of our clients are university graduated, this information is very important in our marketing campaign why : 
    - Adjusting the message and tone of our camapign by using a professional and intellectual language thta aligns with the education level 
    - Highlighting facts, and exerts opinions while using logical and well structured arguments 
    - Using the right channels for displaying our camapaing, university graduates are likely to consume information from credible  sources and professional platforms 
    - Use logical & Vlaue-Based Pricing Strategies by indicating the ROI, the cost-benefit analysis... 

In the other hand we noticed in our analysis that the majority are Married or in a relationship, in this situation we can adjust our marketing stratefy to appeal to couples interests, emotions, and decision-making behavior by : 
    - Emphasizing shared experiences and relationships 
    - adjusting our ad copy and visuals

The impact of these insights on our marketing strategy varies depending on the industry, the companies objectives, the product... 


```python
# Can we see any correlation in our customers demographics ? 

selected_columns = ["Age", "Income","marital_Divorced",  "marital_Married", "marital_Single", "marital_Together", "marital_Widow",'education_2n Cycle',
                    'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD', 'Kidhome', 'Teenhome'  ]
df_selected = data[selected_columns]

corr_matrix = df_selected.corr()
plt.figure(figsize=(9,9))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix ")
plt.show()
```

We can clearly notice that our demographical features are not highly correlated but we can still get some insights : 
    - there is a positive correlation between age and income ( not very high)
    - we notice also a correlation between age and marital_widow
    - there is also a correlation between age and teenhome 
    - and a negative correlation between income and kidhome

# Sales and Behavior Analysis

```python
# in this part we will analysie the purchasing behavior of our clients 

# what is the distribution of each product on sales 
products = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]
total_sales = data[products].sum()

# lets plot the distribution 
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Distribution of Sales Across Different Products")
plt.show()
```

As we notice in this graphe our clients spend more in Wines with 50.5 % of sales, followed by meat prodcts with 27,2% of sales 

```python
# Wat is the part of online purchases and store purchases 

web_purchases = data['NumWebPurchases'].sum()
store_purchases = data['NumStorePurchases'].sum()
catalog_purchases = data['NumCatalogPurchases'].sum()


# Labels and Values 
label = ['Online Purchases', 'In_store Purchases', 'Catalogue purchases']
sizes = [web_purchases, store_purchases, catalog_purchases]
colors = ['#66b3ff', '#ff9999', '#ffc300' ]

# Plot 
fig, ax = plt.subplots(figsize=(6,6))
ax.pie(sizes, labels=label, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'edgecolor':'black'}, pctdistance=0.85)
ax.add_artist(plt.Circle((0,0), 0.50, fc='white'))
plt.title("Share of Online vs In Store Purchases")
plt.show()

```

```python
#now lets see if these spending have a relationship with the marital situation and the home composition of our client

selected_columns = ["Age", "Income", 'Kidhome', 'Teenhome',"MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]
df_selected = data[selected_columns]

corr_matrix = df_selected.corr()
sns.heatmap(data= corr_matrix,annot=True, cmap="coolwarm", linewidths=0.5 )

plt.show()
```

```python
# what is the distribution of expenses (MntTotal) for Income and Age and the home type 
# create the subplot 
fig,(ax1, ax2) = plt.subplots(1,2, figsize=(18,6))

# plot the relation between income and expenses 
sns.scatterplot(data= data, x= data['Income'], y= data['MntTotal'], ax=ax1, hue=data['Income'], palette='rocket_r')
ax1.set_title('The relationship between the expenses and income')

# plot the relation between age and expenses 
sns.scatterplot(data=data, x=data['Age'], y=data['MntTotal'], ax=ax2, hue=data['Age'], palette='viridis' )
ax2.set_title('The relationship between the expenses and age')

plt.show()


```
